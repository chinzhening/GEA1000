```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import data/processed.csv
data = pd.read_csv("data/processed.csv")
```

```python
def encode_smoking_habits(data: pd.DataFrame) -> pd.DataFrame:
    # 1 to 5, do x >= 3 as non-smoker
    df = data.copy()
    df["Smoking_Habits_Encoded"] = df["Smoking_Habits"].apply(lambda x: "Smoker" if x < 3 else "Non-Smoker")
    return df

data = encode_smoking_habits(data)
```

```python
# Boxplot of smoking vs. sleep level
plt.figure(figsize=(8, 6))
sns.boxplot(x="Smoking_Habits_Encoded", y="Hours_Slept", data=data, color="#444444")
plt.title("Boxplot of Smoking Habits vs. Hours Slept")
plt.xlabel("Smoking Habits")
plt.ylabel("Hours Slept")
plt.show()
```
